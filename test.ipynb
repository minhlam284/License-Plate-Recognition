```python
import os
import cv2

# Đường dẫn tới thư mục cần kiểm tra
input_folder = "/Users/kaiser_1/Downloads/Anh bien số/13"

# Lấy danh sách các tệp trong thư mục
image_files = [f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))]

# Đếm số tệp lỗi
error_count = 0

# Kiểm tra từng tệp
for img_name in image_files:
    img_path = os.path.join(input_folder, img_name)
    try:
        img = cv2.imread(img_path)
        if img is None:
            print(f"Lỗi đọc tệp: {img_name}")
            os.remove(img_path)
            print(f"Đã xóa tệp lỗi: {img_name}")
            error_count += 1
    except Exception as e:
        print(f"Lỗi khi mở tệp {img_name}: {e}")
        os.remove(img_path)
        print(f"Đã xóa tệp lỗi: {img_name}")
        error_count += 1

print(f"Tổng số tệp lỗi đã bị xóa: {error_count}")
```
